```python
def nodes_down_unwrap(freqs_full, freqs, vis):
    freqs_inc, vis_inc = freqs[::-1], vis[::-1]
    interp_pts = np.setdiff1d(freqs_full[::-1], freqs_inc, assume_unique=True)
    bin_inds = np.digitize(interp_pts, freqs_inc)
    phase_diff = np.angle(vis_inc[bin_inds]/vis_inc[bin_inds-1])/(freqs_inc[bin_inds]-freqs_inc[bin_inds-1])
    vis_to_add = np.exp(1j*(phase_diff*(interp_pts-freqs_inc[bin_inds-1]) + np.pi))*vis_inc[bin_inds-1]
    vis = np.insert(vis_inc, bin_inds, vis_to_add)
    phase = np.unwrap(np.angle(vis[::-1]))
    condition = [freq in freqs for freq in freqs_full]
    return phase[np.where(condition)]
```

```python
figure(figsize(18,6))
t_index = -1 # Last moment of the acquisition
xlabel('Frequency (MHz)', size=20)
ylabel('rad', size=20)
title('Phase Spectra', size=20)
for i in cross_list: # Loop through cable channels
    phase = nodes_down_unwrap(freqs_full, freqs, vis[:,i,t_index]) # Unwrap phase in frequency
    plot(freqs, phase, label=str(inprds[i][0])) # Plot the phase spectrum. Should be completely linear
    m, b = np.polyfit(freqs*1.e6, phase/(2*np.pi), 1) # Fit a line to the spectrum
    print str(inprds[i][0]) + ': %.3f' %(abs(m)*0.85*3.e8) + 'm' # Infer a cable length from the slope
legend(loc='best', numpoints=1, fancybox=True)
```

```python
# define fitting function
def quadratic(x, *p):
    return p[0]*x*x + p[1]*x + p[2]
```

```python
# Example of fit for one cable and frequency
i = 4
j = 447
delay_frac = (delays[:,i] - np.mean(delays[:,i]))/np.mean(delays[:,i]) # calculate fractional changes
gain_frac = (lin_gains[:,i,j] - np.mean(lin_gains[:,i,j]))/np.mean(lin_gains[:,i,j])
par, cov = curve_fit(quadratic, delay_frac, gain_frac, p0=(1.,1.,1.)) # perform fit. curve_fit has arguments
# (function, x, y, p0= initial guess)
print par
fit = quadratic(delay_frac, *par) # Get the y values of the fitted curve by plugging in the x values and the fitted parameters
figure(figsize(18,6))
xlabel('$\Delta D/\\bar{D}$', size=20)
ylabel('$\Delta G/\\bar{G}$', size=20)
title(str(freqs[j]) + 'MHz', size=20)
plot(delay_frac, gain_frac, 'mo', label=str(cable_ids[i])) # plot scatter points
plot(delay_frac, fit, color='deepskyblue', label='Quadratic Fit') # plot fit
legend(loc='best', numpoints=1, fancybox=True)
ticklabel_format(style='sci', axis='y', scilimits=(0,0))
ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid('on')
```

```python
# Loop through all cables and frequencies, fitting for the coefficients and saving them
alpha = np.zeros((len(cable_ids),len(freqs),3))
for i in range(len(cable_ids)):
    delay_frac = (delays[:,i] - np.mean(delays[:,i]))/np.mean(delays[:,i])
    for j in range(len(freqs)):
        gain_frac = (lin_gains[:,i,j] - np.mean(lin_gains[:,i,j]))/np.mean(lin_gains[:,i,j])
        par, cov = curve_fit(quadratic, delay_frac, gain_frac, p0=(-5.e4,50.,5.e-3))
        #print par
        alpha[i,j,:] = par
np.save('test_3_results/alpha.npy', alpha)
```
