### N.B. Only crates 4-7 in this dataset

```python
c, s, i = (4, 5, 0)
input_slice = np.where(np.logical_not((raw_data['slot'][:] != s) + (raw_data['crate'][:] != c) 
                                      + (raw_data['adc_input'][:] != i)))[0]
```

### N.B. Must cast as float to add these

```python
plt.plot(raw_data['timestamp'][input_slice,:], 
         np.log2(np.sqrt(np.mean(np.square(raw_data['timestream'][input_slice,:], dtype=float), axis=1))),
        '.')
```

```python
plt.hist(raw_data['timestream'][input_slice,:].flatten(), bins=30, range=(-60, 60));
plt.title("Crate {:d}, Slot {:d}, Input {:d} for {:d} frames".format(c, s, i, len(input_slice)))
```
